```python
print("CALCULADORA DE BASES")
print("--------------------------------------------------------")
print("ALUNO: SIDICLEI NECKEL")
print("BSI 1")
print( )
# transformar decimal em binario
def decimal_para_binario(decimal):
    binario = ""
    if decimal == 0:
        binario = "0"
    else:
        while decimal > 0:
            digito = decimal % 2
            binario = str(digito) + binario
            decimal //= 2
    return binario
# fim transformacao decimal em binario

# transformar decimal em octal
def decimal_para_octal(decimal):
    octal = ""
    if decimal == 0:
        octal = "0"
    else:
        while decimal > 0:
            resto = decimal % 8
            octal = str(resto) + octal
            decimal //= 8
    return octal
# fim transformacao decimal em octal

# transformar decimal em hexadecimal
def decimal_para_hexadecimal(decimal):
    hexadecimal = ""
    if decimal == 0:
        hexadecimal = "0"
    else:
        digitos_hexa = "0123456789ABCDEF"
        while decimal > 0:
            resto = decimal % 16
            hexadecimal = digitos_hexa[resto] + hexadecimal
            decimal //= 16
    return hexadecimal
# fim transformacao decimal em hexadecimal

# transformar binario em decimal
def binario_para_decimal(binario):
    decimal = 0
    potencia = 0
    while binario != 0:
        digito = binario % 10
        decimal += digito * (2 ** potencia)
        binario //= 10
        potencia += 1
    return decimal
# fim transformacao binario em decimal

# transformar octal em decimal
def octal_para_decimal(octal):
    decimal = 0
    potencia = 0
    while octal != 0:
        digito = octal % 10
        decimal += digito * (8 ** potencia)
        octal //= 10
        potencia += 1
    return decimal
# fim transformacao octal em decimal

# transformar hexadecimal em decimal
def hexadecimal_para_decimal(hexadecimal):
    decimal = 0
    base = 16
    for digit in hexadecimal:
        if digit.isalpha():
            digit = ord(digit.upper()) - ord('A') + 10
        else:
            digit = int(digit)
        decimal = decimal * base + digit
    return decimal
# fim transformacao hexadecimal em decimal


print("Calculadora de bases")
print("Escolha uma opção para a base:")
print("1 = decimal")
print("2 = binário")
print("3 = octal")
print("4 = hexadecimal")

opcao = input("Insira uma opção: ")


print("Calculadora de bases")
print("Escolha uma opção para a base de saida:")
print("A a base de saída deve ser diferente da base de entrada!!")
print("5 = decimal")
print("6 = binário")
print("7 = octal")
print("8 = hexadecimal")

opcaos = input("Insira uma opção: ")


if opcao == "1":
    decimal = int(input("Insira um número decimal: "))
    binario = decimal_para_binario(decimal)
    octal = decimal_para_octal(decimal)
    hexadecimal = decimal_para_hexadecimal(decimal)
    if opcaos == "6":
        print("Binário:", binario)
    if opcaos == "7":
        print("Octal:", octal)
    if opcaos == "8":
        print("Hexadecimal:", hexadecimal)
    if opcaos == "5":
        print (" O número em questão já está nesta base")

elif opcao == "2":
    binario = int(input("Insira um número binário: "))
    decimal = binario_para_decimal(binario)
    octal = decimal_para_octal(decimal)
    hexadecimal = decimal_para_hexadecimal(decimal)
    if opcaos == "5":
        print("Decimal:", decimal)
    if opcaos == "7":
        print("Octal:", octal)
    if opcaos == "8":
        print("Hexadecimal:", hexadecimal)
    if opcaos == "6":
        print (" O número em questão já está nesta base")


elif opcao == "3":
    octal = int(input("Insira um número octal: "))
    decimal = octal_para_decimal(octal)
    binario = decimal_para_binario(decimal)
    hexadecimal = decimal_para_hexadecimal(decimal)
    if opcaos == "5":
        print("Decimal:", decimal)
    if opcaos == "6":
        print("Binário:", binario)
    if opcaos == "8":
        print("Hexadecimal:", hexadecimal)
    if opcaos == "7":
        print (" O número em questão já está nesta base")

elif opcao == "4":
    hexadecimal = input("Insira um número hexadecimal: ")
    decimal = hexadecimal_para_decimal(hexadecimal)
    binario = decimal_para_binario(decimal)
    octal = decimal_para_octal(decimal)
    if opcaos == "5":
        print("Decimal:", decimal)
    if opcaos == "6":
        print("Binário:", binario)
    if opcaos == "7":
        print("Octal:", octal)
    if opcaos == "8":
        print (" O número em questão já está nesta base")

else:
    print("Opção inválida. Por favor, escolha uma opção válida.")
print ( )
print ( )
print ( )
print ( )
print ("           FIM DO PROGRAMA             ")
print ("GRATO POR TER TESTADO ESSE CÓDIGO, OBRIGADO!. ")
```
